```python
from crewai import Crew, Task, Agent
```

```python
from dotenv import load_dotenv

load_dotenv()

# # Importing the necessary libraries
# import os
```

```python
sales_rep_agent = Agent(
    role="Sales Representative",
    goal=("Identify high-value leads that match"
        "our Ideal Customer profile"),
    backstory=(
        "As a part of the dynamic sales team at AbbasiKoders, your mission is scour the digital landscape for potential leads."
        "Armed with cutting-edge tools & a strategic mindset, you analyze data, trends & interactions to unearth opportunities that others might overlook."
        "Your works is crucial in paving the way for meaningful engagments & driving the company's growth"
    ),
    allow_delegation=False,
    verbose=True
)
```

```python
lead_sales_rep_agent=Agent(
    role="Lead Sales Representative",
    goal="Nuture Lead with Personalized, compelling communications",
    backstory=(
    "Within the Vibrant ecosystem of AbbasiKoder's sales department, you Standout as the bridge between potential clients & the solutions they need."
    "By Creating engaging, personalized messages, you not only inform leads about our offerings but also make them feet seen & heard."
    "Your role is pivotal in converting interest into action, guiding leads through the journey from curiosity to commitment."
    ),
    allow_delegation=False,
    verbose=True
)
```

```python
from crewai_tools import DirectoryReadTool, FileReadTool, SerperDevTool
```

```python
directory_read_tool = DirectoryReadTool("./instructions")
file_read_tool = FileReadTool()
search_tool = SerperDevTool()
```

```python
from crewai.tools import BaseTool
```

```python
from typing import Any


class SentimentAnalysisTool(BaseTool):
    name: str = "Sentiment Analysis Tool"
    description: str = ("Analyzed the sentiment of text"
                        "to ensure positive & engaging communication."
                        )
    
    def _run(self, text:str) -> str:
        return "positive"
```

```python
sentiment_analysis_tool = SentimentAnalysisTool()
```

```python
lead_profiling_task=Task(
    description=(
    "Conduct an in-depth analysis of {lead_name}, a company in the {industry} sector that recently showed interest in our solutions."
    "Utilize all available data sources to compile a detailed profile, focusing on key decision-makers, recent business developments & potentials needs that align with our offerings."
    "This task is crucial for tailoring our enagagement Strategy effectively, Don've make assumptions & only information you absolutely sure about."
    ),
  expected_output=(
    "A comprehensive report on {lead_name}, including company background, Key personnel, recent milestones & identified needs."
    "Highlight potentials areas where our solutions can provide value & suggest personalized enagagement strategies."
  ),
  agent=sales_rep_agent,
  tools=[directory_read_tool, file_read_tool, search_tool],
)
```

```python
personalized_outreach_task=Task(
    description=(
    "Using the insights gathered from the lead profiling report on {lead_name}, Craft a personalized outreach campaign aimed at {key_decision_maker}, the {position} of {lead_name}."
    "The campaign should address their recent {milestone} & how our solutions can support their goals."
    "Your Communication mustt resonate with {lead_name}'s company culture & values, demonstrating a deep understanding of their business & needs, Don't make assumptions & only use information you absolutely sure about."
    ),
  expected_output=(
    "A series of personalized email drafts tailored to {lead_name}, specifically targeting {key_decision_maker}."
    "Each draft should include a compelling narrative that connects our solutions with their recent achievments & future goals."
    "Ensure the tone is engaging, profesional & Aligned wit {lead_name}'s corporate identity."
    ),
  agent=lead_sales_rep_agent,
  tools=[sentiment_analysis_tool, search_tool],
  output_file="customer_outreach.md"
)
```

```python
crew = Crew(
    agents=[sales_rep_agent, lead_sales_rep_agent],
    tasks=[lead_profiling_task, personalized_outreach_task],
    verbose=True,
    memory=True,
)
```

```python
inputs = {
    "lead_name": "DeepLearningAI",
    "industry": "Online Learning Platform",
    "key_decision_maker": "Andrew Ng",
    "position": "CEO",
    "milestone": "product launch"
}

result = crew.kickoff(inputs=inputs)
```
